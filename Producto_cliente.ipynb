```python
estado = df.Estado
estado.value_counts(dropna=False).plot(kind='bar');
estado_ordenado = estado.value_counts(dropna=True).sort_values(ascending=False)
for i in range(len(estado_ordenado)):
    if(estado_ordenado.index[i] == 'Utah'):
        print(estado_ordenado.index[i], ' \t\t| Porcentaje: ',format(estado_ordenado.values[i] / estado_ordenado.sum() * 100, '.2f'), '%', '\t| Cantidad: ', estado_ordenado.values[i])
    else:
        print(estado_ordenado.index[i], ' \t| Porcentaje: ',format(estado_ordenado.values[i] / estado_ordenado.sum() * 100, '.2f'), '%', '\t| Cantidad: ', estado_ordenado.values[i])
```

```python
# Analizaremos las variables de los tres estados mas representativos de nuestro dataset
df_estados = df[(df.Estado == "California") | (df.Estado == "Nevada") | (df.Estado == "Arizona")]
genero_estados = df_estados.Género.value_counts( dropna=False)
genero_estados.plot(kind='pie', title='Género en California, Nevada y Arizona', figsize=(10,5), fontsize=20)
for i in range(len(genero_estados)):
  print(genero_estados.index[i], '\t| Porcentaje: ' , format(genero_estados.values[i]/len(df_estados)*100, '.2f') , '%', '| Cantidad: ' , genero_estados.values[i])
```

```python
# Edades por genero en los estados que mas se compran propiedades
df_estados[(df_estados.Género == 'M')] # Hombres [Male]
df_estados[(df_estados.Género == 'F')] # Mujeres [Female]
df_estados[(df_estados.Género != 'M') & (df_estados.Género != 'F')] # Otros [Nan], no se puede utilizar pues no tienen valores en los intervalos
# Hombres [Male]
df_estados[(df_estados.Género == 'M')].Intervalo.value_counts().plot(kind='bar', xlabel='Edad', ylabel='Cantidad', title='Intervalos de edades para hombres')
```

```python
# Mujeres [Female]
df_estados[(df_estados.Género == 'F')].Intervalo.value_counts().plot(kind='bar', xlabel='Edad', ylabel='Cantidad', title='Intervalos de edades para mujeres')
```

```python
# Propósitos por los que mas se vendieron 
df_estados.Propósito.value_counts().plot(kind='pie', figsize=(10,5), fontsize=12, title='Propósito por el que se vendieron más propiedades en los estados mas representativos')
```

```python
df_estados[(df_estados.Propósito == 'Hogar')].Género.value_counts().plot(kind='pie', figsize=(10,5), title='Género que mas compro propiedades para tener un Hogar')
```

```python
df_estados[(df_estados.Propósito == 'Inversión')].Género.value_counts().plot(kind='pie', figsize=(10,5), title='Género que mas compro propiedades para Invertir')
```
