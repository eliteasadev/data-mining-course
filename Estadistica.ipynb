```python
import matplotlib.pyplot as plt
import numpy as np
import statistics as sts
```

```python
# Cargar datos desde un archivo CSV
arg1 = np.loadtxt('./data.csv', delimiter=',', dtype=np.float32)
# Mostrar la forma (dimensiones) del array cargado
print(arg1.shape)

# Cambiar la forma del array a 1 fila y 30 columnas
arg2 = arg1.reshape(1, 30)  # dimensión 1, límite 30
# Mostrar la nueva forma del array
print(arg2.shape)

# Calcular y mostrar la altura más alta en el array
altura_maxima = np.max(arg1)
print(f"Altura más alta: {altura_maxima:.2f} metros")

# Calcular y mostrar la altura más baja en el array
altura_minima = np.min(arg1)
print(f"Altura más baja: {altura_minima:.2f} metros")

# Calcular y mostrar el rango de alturas (máxima - mínima)
rango_alturas = altura_maxima - altura_minima
print(f"Rango de alturas: {rango_alturas:.2f} metros")

# Ordenar las alturas en orden ascendente
arg2 = np.sort(arg1)

# Mostrar el array ordenado con un mensaje descriptivo
print("Alturas ordenadas en orden ascendente:")
for i, altura in enumerate(arg2, start=1):
    print(f"{i}. {altura:.2f} metros")

# Mostrar las 5 alturas más altas (últimos 5 elementos del array ordenado)
print("Los 5 estudiantes más altos:")
for i, altura in enumerate(arg2[-5:], start=1):
    print(f"{i}. {altura:.2f} metros")

# Mostrar las 5 alturas más bajas (primeros 5 elementos del array ordenado)
print("\nLos 5 estudiantes más bajos:")
for i, altura in enumerate(arg2[0:5], start=1):
    print(f"{i}. {altura:.2f} metros")

# Contar cuántos estudiantes tienen altura mayor o igual a 1.25 metros
suma1_25 = np.where(arg2 >= 1.25, 1, 0).sum()
print(f"Número de estudiantes con altura mayor o igual a 1.25 metros: {suma1_25}")

# Contar cuántos estudiantes tienen altura menor a 1.28 metros
suma1_28 = np.where(arg2 < 1.28, 1, 0).sum()
print(f"Número de estudiantes con altura menor a 1.28 metros: {suma1_28}")

# Calcular el porcentaje de estudiantes cuya altura está entre 1.24 y 1.28 metros (incluyendo 1.24 y excluyendo 1.28)
suma1_24_1_28 = np.where((arg2 >= 1.24) & (arg2 < 1.28), 1, 0).sum() / 30
# Mostrar el porcentaje redondeado a 2 decimales
porcentaje = np.round(suma1_24_1_28 * 100, 2)
print(f"Porcentaje de estudiantes con altura entre 1.24 y 1.28 metros: {porcentaje:.2f}%")
```

```python
# Crear una figura y un eje para el histograma
fig, ax = plt.subplots()

# Encontrar los valores únicos y sus frecuencias en el array ordenado
variables, repetidas = np.unique(arg2, return_counts=True)

# Imprimir los valores únicos y sus frecuencias
print("Valores únicos y sus frecuencias:")
for valor, frecuencia in zip(variables, repetidas):
    print(f"{valor:.2f}: {frecuencia}")

# Calcular las frecuencias absolutas simples
frecuencias_absolutas_simples = repetidas

# Calcular las frecuencias absolutas acumuladas
frecuencias_absolutas_acumuladas = np.cumsum(frecuencias_absolutas_simples)

# Calcular las frecuencias relativas simples
frecuencias_relativas_simples = np.round((frecuencias_absolutas_simples / 30) * 100, 2)

# Calcular las frecuencias relativas acumuladas
frecuencias_relativas_acumuladas = np.cumsum(frecuencias_relativas_simples)

# Imprimir las frecuencias
print("\nFrecuencias absolutas simples:")
for valor, frecuencia in zip(variables, frecuencias_absolutas_simples):
    print(f"{valor:.2f}: {frecuencia}")

print("\nFrecuencias absolutas acumuladas:")
for valor, frecuencia in zip(variables, frecuencias_absolutas_acumuladas):
    print(f"{valor:.2f}: {frecuencia}")

print("\nFrecuencias relativas simples (%):")
for valor, frecuencia in zip(variables, frecuencias_relativas_simples):
     print(f"{valor:.2f}: {frecuencia:.2f}")

print("\nFrecuencias relativas acumuladas (%):")
for valor, frecuencia in zip(variables, frecuencias_relativas_acumuladas):
    print(f"{valor:.2f}: {frecuencia:.2f}")

# Crear los límites de los bins para el histograma
b = np.append(variables, 1.31)

# Crear el histograma
frecuencia, _, _ = ax.hist(arg2, bins=b, linewidth=0.5, edgecolor="white")

# Configurar el eje x
ax.set(xlim=(1.19, 1.32), xticks=np.arange(1.2, 1.32, .01),
       xlabel="Altura (metros)")

# Configurar el eje y
ax.set(ylim=(1, 7), yticks=np.linspace(0, 6, 7, dtype=np.int16),
       ylabel="Frecuencia")

# Agregar una cuadrícula
ax.grid()

# Mostrar el gráfico
plt.show()
```
