```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

#make data
#genera un numero
np.random.seed(1)
#una lista generado de datos entre 0 y 1.5 con una cantidad de 200 datos y el 4, todo esto se guarda en x
x=4+np.random.normal(0,1.5,200)

#plot
fig, ax=plt.subplots()

#bins es la clase
ax.hist(x, bins=8, linewidth=0.5, edgecolor='white')

#xticks acaba en n-1
ax.set(xlim=(0,8), xticks=np.arange(1,8),
       #El 9 equivale a las divisiones de y
       ylim=(0,56), yticks=np.linspace(0,56,9))

plt.show()
```

```python
rnd=np.random.default_rng(2456)
x=rnd.integers(-5,5,size=15)
y=rnd.integers(-5,5,size=15)

fig,ax=plt.subplots()
ax.scatter(x,y)
ax.grid(True)
```
