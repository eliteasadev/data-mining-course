```python
import matplotlib.pyplot as plt
import numpy as np
import statistics as sts
```

```python
# Cargar datos desde un archivo CSV en un array de tipo float32
arg1 = np.loadtxt('./data3.csv', delimiter=',', dtype=np.float32)

# Imprimir la forma (dimensiones) del array cargado
print("Dimensiones del array cargado: ", arg1.shape)

# Reorganizar el array en una forma de 50 filas y 10 columnas
arg2 = arg1.reshape(50, 10)

# Imprimir la nueva forma del array
print("Dimensiones del array reestructurado: ", arg2.shape)

# Calcular y mostrar el número más alto en el array
print("Número más alto en el conjunto de datos: ", np.max(arg1))

# Calcular y mostrar el número más bajo en el array
print("Número más bajo en el conjunto de datos: ", np.min(arg1))

# Calcular y mostrar el rango de los números en el array
rango = np.max(arg1) - np.min(arg1)
print("Rango de los números (máximo - mínimo): ", rango)

# Ordenar los elementos del array en un solo vector
arg2 = np.sort(arg1, axis=None)

# Imprimir todos los elementos ordenados
print("Elementos ordenados: ", arg2)

# Mostrar los 5 números más altos
print("Los 5 números más altos: ", arg2[-5:])

# Mostrar los 5 números más bajos
print("Los 5 números más bajos: ", arg2[:5])

# Contar cuántos números son mayores o iguales a 22.5
suma1 = np.where(arg2 >= 22.5, 1, 0).sum()
print("Cantidad de números mayores o iguales a 22.5: ", suma1)

# Contar cuántos números son menores que 25
suma2 = np.where(arg2 < 25, 1, 0).sum()
print("Cantidad de números menores que 25: ", suma2)

# Calcular el porcentaje de números en el rango [22.4, 25)
suma1_2 = np.where((arg2 >= 22.4) & (arg2 < 25), 1, 0).sum() / 500
print("Porcentaje de números en el rango [22.4, 25): ", np.round(suma1_2 * 100, 2), "%")

# Calcular el ancho de clase utilizando la regla de Sturges
ancho = 1 + (3.3 * (np.log10(500)))
print("Ancho de clase (regla de Sturges): ", ancho)

# Calcular el intervalo de clase
intervalo = (np.max(arg1) - np.min(arg1)) / ancho
print("Intervalo de clase: ", intervalo)
```

```python
# Crear una figura y un eje para el gráfico
fig, ax = plt.subplots()

# Obtener los valores únicos y sus frecuencias en el array ordenado
variables, repetidas = np.unique(arg2, return_counts=True)

# Imprimir los valores únicos y sus frecuencias
print("Valores únicos y sus frecuencias:")
for var, rep in zip(variables, repetidas):
    print(f"Valor: {var:.2f}, Frecuencia: {rep}")

# Frecuencias absolutas simples son las repeticiones de cada valor
frecuencias_absolutas_simples = repetidas

# Calcular las frecuencias absolutas acumuladas
frecuencias_absolutas_acumuladas = np.cumsum(frecuencias_absolutas_simples)

# Calcular las frecuencias relativas simples como porcentaje
frecuencias_relativas_simple = np.round((frecuencias_absolutas_simples / 500) * 100, 2)

# Calcular las frecuencias relativas acumuladas
frecuencias_relativas_acumuladas = np.cumsum(frecuencias_relativas_simple)

# Imprimir las frecuencias absolutas y relativas
print("\nFrecuencias absolutas simples:")
for freq in frecuencias_absolutas_simples:
    print(freq)

print("\nFrecuencias absolutas acumuladas:")
for freq in frecuencias_absolutas_acumuladas:
    print(freq)

print("\nFrecuencias relativas simples (%):")
for freq in frecuencias_relativas_simple:
    print(freq)

print("\nFrecuencias relativas acumuladas (%):")
for freq in frecuencias_relativas_acumuladas:
    print(freq)

# Añadir un límite superior al rango de datos para el histograma
b = np.append(variables, 51)

# Crear el histograma con los datos y los límites de los bins
frecuencia = ax.hist(arg2, bins=b, linewidth=1.0, edgecolor="white")

# Configurar los límites y etiquetas de los ejes
ax.set(xlim=(0.29, 52), 
       xticks=np.arange(0.3, 52, intervalo),
       ylim=(1, 6), 
       yticks=np.linspace(0, 5, 6, dtype=np.int16))

# Añadir una cuadrícula al gráfico
ax.grid()

# Mostrar el gráfico
plt.show()
```

```python
# Calcular y mostrar la media aritmética de los datos no agrupados
media_aritmetica = np.sum(arg2) / 500
print("Media aritmética: {:.2f}".format(media_aritmetica))

# Calcular y mostrar la media geométrica
media_geometrica = sts.geometric_mean(arg2)
print("Media geométrica: {:.2f}".format(media_geometrica))

# Calcular y mostrar la media armónica
media_armonica = sts.harmonic_mean(arg2)  # Corregido para usar la función correcta
print("Media armónica: {:.2f}".format(media_armonica))

# Calcular y mostrar la mediana
mediana = np.median(arg2)
print("Mediana: {:.2f}".format(mediana))

# Calcular y mostrar las múltiples modas (distribución bimodal)
modas = sts.multimode(arg2)
print("Múltiples modas: ", modas)

# Calcular y mostrar los cuartiles
cuartiles = [round(q, 2) for q in sts.quantiles(arg2, n=4)]
print("Cuartiles: ", cuartiles)

# Calcular y mostrar los deciles
deciles = [round(q, 2) for q in sts.quantiles(arg2, n=10)]
print("Deciles: ", deciles)

# Calcular y mostrar los percentiles
percentiles = [round(q, 2) for q in sts.quantiles(arg2, n=100)]
print("Percentiles: ", percentiles)
```
